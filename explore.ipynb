# Double-integrator

\begin{cases}
  \dot{x_1} &= u_p + E (1 - u_l) + O - R u_r x_1 \\
  \dot{x_2} &= \alpha x_1 - R u_r x_2
\end{cases}

to simulate neuron saturation:

\begin{cases}
  \dot{x_1} &= u_p + E (1 - u_l) + O - R u_r x_1 + \min(0, |x_1| - 1.0) \\
  \dot{x_2} &= \alpha x_1 - R u_r x_2  + \min(0, |x_2| - 1.0)
\end{cases}

```python
import numpy as np
import matplotlib.pyplot as plt
import PyDSTool as dst
```

```python
seq = ['C']
tlen = 10
tdomain = [0, len(seq) * tlen]
dt = 0.01
time = np.arange(tdomain[0], tdomain[1]+dt, dt)
press = np.zeros_like(time)
lights = np.ones_like(time)
reward = np.zeros_like(time)

for ix, trial in enumerate(seq):
    start = ix * tlen
    press[np.logical_and(time > start + 0.8, time <= start + 1.0)] = 5
    if trial.lower() == 'c':
        reward[np.logical_and(time > start + 2.6, time <= start + 4.0)] = 1
    elif trial.lower() == 'p':
        lights[np.logical_and(time > start + 1.3, time <= start + 2.7)] = 0
    elif trial.lower() == 'l':
        lights[np.logical_and(time > start + 2.6, time <= start + 4.0)] = 0
```

```python
# Make input
u = dst.makeDataDict(['u_p', 'u_l', 'u_r'], [press, lights, reward])
dsargs = dst.args(name='piecewise')
dsargs.tdata = time
dsargs.ics = u
dsargs.method = 'constant'
utable = dst.Generator.ExtrapolateTable(dsargs)
utabletraj = utable.compute('piecewise')
udata = utabletraj.sample(dt=0.01)

# These lines are slightly offset in order to see them all at once
plt.plot(udata['t'], udata['u_p']+0.01, lw=2, label="Pressing")
plt.plot(udata['t'], udata['u_l'], lw=2, label="Lights")
plt.plot(udata['t'], udata['u_r']-0.01, lw=2, label="Reward")
plt.ylim(-.1, 1.1)
plt.legend(frameon=False, loc='center right')
```

```python
amp = dst.Par(0.0, 'amp')
freq = dst.Par(0.185, 'freq')
phase = dst.Par(1.3, 'phase')

dsargs = dst.args(name='oscillation')
dsargs.tdomain = tdomain
dsargs.pars = [amp, freq, phase]
dsargs.varspecs = {'O': "amp * sin(2 * pi * freq * globalindepvar(t) + phase)"}
o_gen = dst.ExplicitFnGen(dsargs)
otraj = o_gen.compute('oscillation')
```

```python
odata = otraj.sample(dt=0.1)
plt.plot(odata['t'], odata['O'], lw=2)
```

```python
# Params
E = dst.Par(1.5, 'err')
R = dst.Par(4, 'R')
alpha = dst.Par(0.7, 'alpha')

dsargs = dst.args(name='doubleintegrator')
dsargs.varspecs = {'x1': 'u_p - err * (1 - u_l) + O - R * u_r * x1 + if(x1 > 1.0, 1.0 - x1, 0) + if (x1 < -1.0, -1.0 - x1, 0)',
                   'x2': 'alpha * x1 - R * u_r * x2 + if(x2 > 1.0, 1.0 - x2, 0) + if (x2 < -1.0, -1.0 - x2, 0)'}
dsargs.vars = ['x1', 'x2']
dsargs.algparams = {'max_step': 0.001}
dsargs.inputs = dict(utabletraj.variables.items() + otraj.variables.items())
dsargs.ics = {'x1': 0., 'x2': 0.}
dsargs.pars = [E, R, alpha]
dsargs.tdata = tdomain
ode = dst.Generator.Vode_ODEsystem(dsargs) # dst.embed( , tdata=[0, 4])
traj = ode.compute('doubleintegrator')
data = traj.sample(dt=0.01)

plt.plot(data['t'], data['x1'], lw=2, label="x_1")
plt.plot(data['t'], data['x2'], lw=2, label="x_2")
# These lines are slightly offset in order to see them all at once
plt.plot(udata['t'], udata['u_p']+0.01, label="Pressing")
plt.plot(udata['t'], udata['u_l'], label="Lights")
plt.plot(udata['t'], udata['u_r']-0.01, label="Reward")
plt.legend(frameon=False, loc='center right')

plt.figure()
plt.plot(data['x2'], data['x1'], lw=2)
```

```python
# Calculate other models' quantities with this model.

# Alexander & Brown
# S, state = x_1
plt.figure()
plt.plot(data['t'], data['x1'])
plt.xlim(right=5)

# V, value = x_2
plt.figure()
plt.plot(data['t'], data['x2'])
plt.xlim(right=5)

# r, reward = r
plt.figure()
plt.plot(udata['t'], udata['u_r'])
plt.xlim(right=5)

# delta not needed, as time course learning is not RPE dependent

# omega_N, unpredicted non-occurrences of reward = max(0, x_2 - r)
plt.figure()
plt.plot(data['t'], np.maximum(np.zeros_like(data['x2']), data['x2'] - udata['u_r'][:-1]))
plt.xlim(right=5)

# omega_P, unpredicted occurrences = max(0, r - x_2)
plt.figure()
plt.plot(data['t'], np.maximum(np.zeros_like(data['x2']), udata['u_r'][:-1] - data['x2']))
plt.xlim(right=5)
```

```python
# Signal going to the T population
release = 1. / (1 + np.exp(-9 * (data['x2'] - 0.9))) * 1.5
plt.plot(data['t'], release)
plt.xlim(right=5)
```

```python
# Effect of changing R
R = np.linspace(0, 4, 21)
x1end = []
x2end = []

for i, rr in enumerate(R):
    ode.set(pars={'R': rr})
    tmp = ode.compute('%3i' % i).sample()
    plt.plot(tmp['x2'], tmp['x1'])
    x1end.append(tmp['x1'][-1])
    x2end.append(tmp['x2'][-1])
ode.set(pars={'R': 4})

plt.figure()
plt.plot(R, x1end, label='$x_1$')
plt.plot(R, x2end, label='$x_2$')
plt.legend(loc='best')
```

```python
# Effect of changing alpha
alpha = np.linspace(0, 1, 21)
x1cue = []
x2cue = []

for i, aa in enumerate(alpha):
    ode.set(pars={'alpha': aa})
    traj = ode.compute('%3i' % i)
    tmp = traj.sample()
    plt.plot(tmp['x2'], tmp['x1'])
    x1cue.append(traj(2.0)['x1'])
    x2cue.append(traj(2.0)['x2'])
ode.set(pars={'alpha': 0.7})

plt.figure()
plt.plot(alpha, x1cue, label='$x_1$')
plt.plot(alpha, x2cue, label='$x_2$')
plt.legend(loc='best')
```

```python
# Effect of changing E
E = np.linspace(0, 2, 21)
x1end = []
x2end = []

for i, err in enumerate(E):
    ode.set(pars={'err': err})
    tmp = ode.compute('%3i' % i).sample()
    plt.plot(tmp['x2'], tmp['x1'])
    x1end.append(tmp['x1'][-1])
    x2end.append(tmp['x2'][-1])
ode.set(pars={'err': 1.5})

plt.figure()
plt.plot(E, x1end, label='$x_1$')
plt.plot(E, x2end, label='$x_2$')
plt.legend(loc='best')
```
