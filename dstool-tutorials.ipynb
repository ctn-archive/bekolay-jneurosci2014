# Harmonic oscillator

$$m \frac{d^2x}{dt^2} = -kx$$

\begin{cases}
  \frac{dy}{dt} &= - \frac{kx}{m}, \\
  \frac{dx}{dt} &= y
\end{cases}

```python
dsargs = dst.args(name='SHM')
dsargs.ics = {'x': 0.4, 'y': 0.4}
dsargs.pars = {'k': 0.3, 'm': 0.5}
dsargs.varspecs = {'x': 'y', 'y': '-k * x / m'}
```

```python
dsargs.tdata = [0, 20]
ode = dst.Generator.Vode_ODEsystem(dsargs)
traj = ode.compute('demo')
pts = traj.sample()

plt.plot(pts['t'], pts['x'], label='x')
plt.plot(pts['t'], pts['y'], label='y')
plt.legend()
plt.xlabel('t')
```

```python
def KE(pts):
    return 0.5 * ode.pars['m'] * pts['y'] ** 2

def PE(pts):
    return 0.5 * ode.pars['k'] * pts['x'] ** 2
```

```python
KE(pts) + PE(pts)
```

# Lorenz Map

$$z(t + 1) = a + b(z(t) - c)^2$$

## Lyapunov exponent

Linearization of Lorenz map along its solutions in log coordinates:

$$zp(t + 1) = zp(t) + \log(\text{abs}(2b(z(t) - c)))$$

Examine the auxiliary value $L = zp(t)/t$

```python
dsargs = dst.args(name='Lorenz map')
dsargs.varspecs = {'z': 'a + b * (z - c) ** 2',
                   'zp': 'zp + log(abs(2 * b * (z - c)))',
                   'L': 'zp / (t + 1)'}
dsargs.pars = {'a': 2.93, 'b': -1.44, 'c': 1.85}
dsargs.vars = ['z', 'zp']  # Implicitly, then, L will be an auxiliary variable
dsargs.ics = {'z': 2, 'zp': 0}
dsargs.ttype = int  # force independent variable type to be integer (discrete time)
```

```python
dsargs.tdomain = [0, 200]
lmap = dst.Generator.MapSystem(dsargs)
traj = lmap.compute('test')
pts = traj.sample()

plt.figure()
plt.plot(pts['t'], pts['z'], 'ko-')
plt.title(lmap.name)
plt.xlabel('t')
plt.ylabel('z')

plt.figure()
plt.plot(pts['t'], pts['L'], 'ro')
plt.title('Lyapunov exponent')
plt.ylabel('L')
plt.xlabel('t')
```

# Symbolic Jacobian for a stiff ODE system

\begin{cases}
  \frac{dy_0}{dt} &= -0.04 y_0 + 10^{-4} y_1 y_2 \\
  \frac{dy_1}{dt} &= 3 \cdot 10^7 y_1^2 \\
  \frac{dy_2}{dt} &= - \frac{dy_0}{dt} - \frac{dy_2}{dt}
\end{cases}

```python
# declarations of symbolic objects
y0 = dst.Var('y0')
y1 = dst.Var('y1')
y2 = dst.Var('y2')
t = dst.Var('t')

# definitions of right-hand sides for each variable, as functions of all variables
ydot0 = dst.Fun(-0.04 * y0 + 1e4 * y1 * y2, [y0, y1, y2], 'ydot0')
ydot2 = dst.Fun(3e7 * y1 * y1, [y0, y1, y2], 'ydot2')
ydot1 = dst.Fun(-ydot0(y0, y1, y2) - ydot2(y0, y1, y2), [y0, y1, y2], 'ydot1')

# the whole vector field as one nonlinear multi-variate function: R^3 -> R^3
F = dst.Fun([ydot0(y0, y1, y2), ydot1(y0, y1, y2), ydot2(y0, y1, y2)], [y0, y1, y2], 'F')

# Diff is the symbolic derivative operator, and returns a QuantSpec object.
# It must be named "Jacobian" for PyDSTool to recognize it as such.
jac = dst.Fun(dst.Diff(F, [y0, y1, y2]), [t, y0, y1, y2], 'Jacobian')
# This defines a function of arguments (t, y0, y1, y2) obtained by differentiating
# F in all three variables, y0, y1, and y2.

# The user could inspect this expression, jac, and make some simplifications
# by hand, to help optimize the speed of its evaluation. Here, we'll
# just use it as is.
dsargs = dst.args(name='jactest', checklevel=2)
dsargs.fnspecs = [jac, ydot0, ydot2]
dsargs.varspecs = {y0: ydot0(y0, y1, y2),
                   y2: ydot2(y0, y1, y2),
                   y1: -ydot0(y0, y1, y2) - ydot2(y0, y1, y2)}
dsargs.ics = {y0: 1.0, y1: 0., y2: 0.}
dsargs.algparams = {'init_step':0.4, 'strictdt': True, 'stiff': True,
                    'rtol': 1e-4, 'atol': [1e-8,1e-14,1e-6]}
dsargs.events = dst.makeZeroCrossEvent(y0 - 0.001, -1, {
    'name': 'thresh_ev', 'eventtol': 10, 'bisectlimit': 20,
    'eventinterval': 500, 'eventdelay': 0,  #otherwise cannot catch event with only one step per run
    'term': True}, [y0])
```

```python
dsargs.tdomain = [0.,1e20]
ode = dst.Generator.Vode_ODEsystem(dsargs)

print "Defined the following internal Python function for Jacobian:"
print ode.funcspec.auxfns['Jacobian'][0]

t0 = 0.

scipy_ode_results = {
    0.0: [1., 0., 0.],
    0.4: [9.85172114e-001,  3.38639538e-005,  1.47940221e-002],
    4.0: [9.05518679e-001,  2.24047569e-005,  9.44589166e-002],
    40.0: [7.15827070e-001,  9.18553479e-006,  2.84163745e-001],
    400.0: [4.50518662e-001,  3.22290136e-006,  5.49478115e-001],
    4000.0: [1.83202242e-001,  8.94237031e-007,  8.16796863e-001],
    40000.0: [3.89833646e-002,  1.62176779e-007,  9.61016473e-001],
    400000.0: [4.93828363e-003,  1.98499788e-008,  9.95061697e-001],
    4000000.0: [5.16819063e-004,  2.06833253e-009,  9.99483179e-001],
    40000000.0: [5.20200703e-005,  2.08090952e-010,  9.99947980e-001],
    400000000.0: [5.21215549e-006,  2.08487297e-011,  9.99994788e-001],
    4000000000.0: [5.25335126e-007,  2.10134087e-012,  9.99999475e-001],
    40000000000.0: [5.63729748e-008,  2.25491907e-013,  9.99999944e-001],
}

for t1 in (0.4 * 10 ** i for i in xrange(0, 12)):
    dt = t1 - t0
    print "\n===============================================\nAt t=",
    print t1, "using dt =", dt
    ode.set(tdata=[t0,t1], algparams={'init_step': dt})
    if t0 > 0.:
        print ode._solver.y
    traj = ode.compute('test', 'c')  # c for continue (acts as f first time)
    try:
        print traj(t1)
        if scipy_ode_results.has_key(t1):
            print 'SciPy:', scipy_ode_results[t1]
            print 'difference:', [traj(t1)[i] - scipy_ode_results[t1][i] for i in xrange(3)]
    except ValueError:
        print "\n----------"
        ode.diagnostics.showWarnings()
        print traj(traj.indepdomain[1])
        print "----------\n"
        t0 = traj.indepdomain[1]
    else:
        t0 = t1
```

# Zero-crossings in the Van der Pol oscillator phase plane

\begin{cases}
  \frac{dx}{dt} &= (y - x^3 - x) / \epsilon \\	   
  \frac{dy}{dt} &= a - x
\end{cases}

where $a > 0$ and $\epsilon > 0$ are parameters.

```python
event_x_a = dst.makeZeroCrossEvent('x-a', 0,
    {'name': 'event_x_a', 'eventtol': 1e-6, 'term': False ,'active': True},
    varnames=['x'], parnames=['a'])

pars = {'eps': 1e-2, 'a': 0.5}
dsargs = dst.args(name='vanderpol')
dsargs.events = [event_x_a]
dsargs.pars = pars
dsargs.algparams = {'max_pts': 3000, 'init_step': 0.02, 'stiff': True}
dsargs.varspecs = {'x': '(y - (x * x * x / 3 - x)) / eps', 'y': 'a - x'}
dsargs.xdomain = {'x': [-2.2, 2.5], 'y': [-2, 2]}
dsargs.fnspecs = {'Jacobian': (['t','x','y'],
                                "[[(1 - x * x) / eps, 1 / eps ], [-1, 0]]")}
dsargs.ics = {'x': pars['a'], 'y': pars['a'] - pars['a'] ** 3 / 3}
```

```python
dsargs.tdata = [0, 3]
vdp = dst.Vode_ODEsystem(dsargs)
traj = vdp.compute('test_traj')
pts = traj.sample()
evs = traj.getEvents('event_x_a')

plt.plot(pts['t'], pts['x'], 'b', linewidth=2)
plt.plot(pts['t'], pts['y'], 'r', linewidth=2)
```

```python
from PyDSTool.Toolbox import phaseplane as pp

# plot vector field
pp.plot_PP_vf(vdp, 'x', 'y', scale_exp=-1)

# only one fixed point, hence [0] at end.
# n=4 uses three starting points in the domain to find any fixed points,
# to an accuracy of eps=1e-8.
fp_coord = pp.find_fixedpoints(vdp, n=4, eps=1e-8)[0]
fp = pp.fixedpoint_2D(vdp, dst.Point(fp_coord), eps=1e-8)

# n=3 uses three starting points in the domain to find nullcline parts, to an
# accuracy of eps=1e-8, and a maximum step for the solver of 0.1 units.
# The fixed point found is also provided to help locate the nullclines.
nulls_x, nulls_y = pp.find_nullclines(vdp, 'x', 'y', n=3, eps=1e-8, max_step=0.1, fps=[fp_coord])

# plot the fixed point
pp.plot_PP_fps(fp)

# plot the nullclines
plt.plot(nulls_x[:,0], nulls_x[:,1], 'b')
plt.plot(nulls_y[:,0], nulls_y[:,1], 'g')

# plot the trajectory
plt.plot(pts['x'], pts['y'], 'k-o', linewidth=2, ms=5)

# plot the event points
plt.plot(evs['x'], evs['y'], 'rs')

plt.axis('tight')
plt.title('Phase plane')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# set integration time to be long enough to settle to the limit cycle,
# and extend y's domain to suit larger eps values
vdp.set(xdomain={'y': [-8, 8]}, #tdata=[0, 30],
        algparams={'init_step': 0.05})

# select 6 values spaced logarithmically between 0.01 and 100
# incrementing the integration time as the oscillations get slower
for eps in np.power(10, np.linspace(-2, 1.25, 6)):
    vdp.set(pars={'eps': eps},
            tdata=[0, vdp.tdata[1]+40])
    traj = vdp.compute('eps_%.4f' % eps)
    pts = traj.sample()
    event_dict = pts.labels.by_label['Event:event_x_a']
    indices = np.sort(event_dict.keys())
    # assume there are at least 3 indices!
    # pick the last two that go through the same point
    ix1, ix2, ix3 = indices[-3:]
    plt.plot(pts['x'][ix1:ix3+1], pts['y'][ix1:ix3+1], label='$\epsilon$=%.3f' % eps)

# find nullclines for new domain (extends the y-nullcline)
nulls_x, nulls_y = pp.find_nullclines(vdp, 'x', 'y', n=3, eps=1e-8, max_step=0.2, fps=[fp_coord])

# plot nullclines again for reference (which are independent of eps)
plt.plot(nulls_x[:,0], nulls_x[:,1], 'k', lw=2)
plt.plot(nulls_y[:,0], nulls_y[:,1], 'k', lw=2)

# plot the fixed point
pp.plot_PP_fps(fp)

plt.axis('tight')
plt.title('Phase portrait as $\epsilon$ varies')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', fontsize='small')
```

# Import NineML models

`NineML` is a markup language for specifying large-scale neural networks, targetting neural simulators like NEURON.

```python
from PyDSTool.Toolbox import NineML
import nineml.abstraction_layer as al

def get_HH_component():
    aliases = [
        "q10 := 3.0**((celsius - 6.3)/10.0)",  # temperature correction factor
        "alpha_m := -0.1*(V+40.0)/(exp(-(V+40.0)/10.0) - 1.0)", # m
        "beta_m := 4.0*exp(-(V+65.0)/18.0)",
        "mtau := 1/(q10*(alpha_m + beta_m))",
        "minf := alpha_m/(alpha_m + beta_m)",
        "alpha_h := 0.07*exp(-(V+65.0)/20.0)", # h
        "beta_h := 1.0/(exp(-(V+35)/10.0) + 1.0)",
        "htau := 1.0/(q10*(alpha_h + beta_h))",
        "hinf := alpha_h/(alpha_h + beta_h)",
        "alpha_n := -0.01*(V+55.0)/(exp(-(V+55.0)/10.0) - 1.0)", # n
        "beta_n := 0.125*exp(-(V+65.0)/80.0)",
        "ntau := 1.0/(q10*(alpha_n + beta_n))",
        "ninf := alpha_n/(alpha_n + beta_n)",
        "gna := gnabar*m*m*m*h",
        "gk := gkbar*n*n*n*n",
        "ina := gna*(ena - V)",  # currents
        "ik := gk*(ek - V)",
        "il := gl*(el - V )"]

    hh_regime = al.Regime(
        "dn/dt = (ninf-n)/ntau",
        "dm/dt = (minf-m)/mtau",
        "dh/dt = (hinf-h)/htau",
        "dV/dt = (ina + ik + il + Isyn)/C",
        transitions=al.On("V > theta", do=al.SpikeOutputEvent())
    )

    # the rest are not "parameters" but aliases, assigned vars, state vars,
    # indep vars, analog_analog_ports, etc.
    parameters = ['el', 'C', 'ek', 'ena', 'gkbar', 'gnabar', 'theta', 'gl', 'celsius']
    analog_ports = [al.SendPort("V"), al.ReducePort("Isyn",reduce_op="+")]
    c1 = al.ComponentClass("HodgkinHuxley",
                           parameters=parameters,
                           regimes=(hh_regime,),
                           aliases=aliases,
                           analog_ports=analog_ports)
    return c1

c = get_HH_component()

# Convert to PyDSTool.ModelSpec and create NonHybridModel object
# Provide extra parameter Isyn which is missing from component definition
# in absence of any synaptic inputs coupled to the model membrane
hhmodel = NineML.get_nineml_model(c, 'HH_9ML', extra_args=[dst.Par('Isyn')])

hhmodel.set(pars={'C': 1.0,
                  'Isyn': 20.0,
                  'celsius': 20.0,
                  'ek': -90,
                  'el': -65,
                  'ena': 80,
                  'gkbar': 30.0,
                  'gl': 0.3,
                  'gnabar': 130.0,
                  'theta': -40.0},
            ics={'V': -70, 'm': 0.1, 'n': 0, 'h': 0.9},
            tdata=[0, 15])
```

```python
hhmodel.compute('test', force=True)
pts = hhmodel.sample('test')
plt.plot(pts['t'], pts['V'], 'k')
plt.title('Hodgkin-Huxley membrane potential')

ev_info = pts.labels.by_label['Event:spikeoutput']
for ev_ix, ev_tdata in ev_info.items():
    plt.plot(ev_tdata['t'], pts[ev_ix]['V'], 'ko')

plt.xlabel('t')
plt.ylabel('V')
```

# 4-D predator-prey system

\begin{cases}
  v1' &= v1*(1-v1/k) - v1*p1/(1+h*v1) \\
  v2' &= v2*(1-v2/k) - v2*p2/(1+h*v2) \\
  p1' &= -1*\mu*p1 + v1*p1/(1+h*v1) + D*(((1+\theta*h*v2)/(1+h*v2))*p2 - ((1+\theta*h*v1)/(1+h*v1))*p1) \\
  p2' &= -1*\mu*p2 + v2*p2/(1+h*v2) + D*(((1+\theta*h*v1)/(1+h*v1))*p1 - ((1+\theta*h*v2)/(1+h*v2))*p2)
\end{cases}

where $D, \mu, \theta, k, h$ are scalar parameters.

```python
# Declare names and initial values for (symbolic) parameters
mu = dst.Par(0.8, 'mu')
k = dst.Par(7, 'k')
D = dst.Par(0.5, 'D')
theta = dst.Par(1, 'theta')
h = dst.Par(0.5, 'h')

# Compute nontrivial boundary equilibrium initial condition from parameters (see reference for derivation)
v1_0 = mu * (2 * D + mu) / (D * (1 - h * mu - theta * h * mu) + mu * (1 - h * mu))
v2_0 = 0.0
p1_0 = (1 + h * v1_0) * (1 - v1_0 / k)
p2_0 = (D / (D + mu)) * (1 + theta * h * v1_0) * (1 - v1_0 / k)

# Declare symbolic variables
v1 = dst.Var('v1')
v2 = dst.Var('v2')
p1 = dst.Var('p1')
p2 = dst.Var('p2')

# Create Symbolic Quantity objects for definitions
v1rhs = v1 * (1 - v1 / k) - v1 * p1 / (1 + h * v1)
v2rhs = v2 * (1 - v2 / k) - v2 * p2 / (1 + h * v2)
p1rhs = -1 * mu * p1 + v1 * p1 / (1 + h * v1) + D * (((1 + theta * h * v2) / (1 + h * v2)) * p2 - ((1 + theta * h * v1) / (1 + h * v1)) * p1)
p2rhs = -1 * mu * p2 + v2 * p2 / (1 + h * v2) + D * (((1 + theta * h * v1) / (1 + h * v1)) * p1 - ((1 + theta * h * v2) / (1 + h * v2)) * p2)

# Build Generator
dsargs = dst.args(name='PredatorPrey')
dsargs.pars = [mu, k, D, theta, h]
dsargs.varspecs = dict(v1=v1rhs,
                       v2=v2rhs,
                       p1=p1rhs,
                       p2=p2rhs)

# Use eval method to get a float value from the symbolic definitions given in
# terms of parameter values
dsargs.ics = dict(v1=v1_0.eval(), v2=v2_0, p1=p1_0.eval(), p2=p2_0.eval())
ode = dst.Generator.Vode_ODEsystem(dsargs)
```

```python
# Hopf bifurcation points
pc = dst.ContClass(ode)

pcargs = dst.args(name='EQ1', type='EP-C')
pcargs.freepars = ['k']
pcargs.StepSize = 1e-2
pcargs.MaxNumPoints = 50
pcargs.MaxStepSize = 1e-1
pcargs.LocBifPoints = 'all'
pcargs.SaveEigen = True
pcargs.verbosity = 2
pc.newCurve(pcargs)

print 'Computing curve...'
start = dst.clock()
pc['EQ1'].forward()
print 'done in %.3f seconds!' % (dst.clock() - start)

pcargs.name = 'HO1'
pcargs.type = 'H-C2'
pcargs.initpoint = 'EQ1:H1'
pcargs.freepars = ['k','D']
pcargs.MaxNumPoints = 50
pcargs.MaxStepSize = 0.1
pcargs.LocBifPoints = ['ZH']
pcargs.SaveEigen = True
pc.newCurve(pcargs)

print 'Computing Hopf curve...'
start = dst.clock()
pc['HO1'].forward()
print 'done in %.3f seconds!' % (dst.clock() - start)

pcargs = dst.args(name='FO1', type='LP-C')
pcargs.initpoint = 'HO1:ZH1'
pcargs.freepars = ['k','D']
pcargs.MaxNumPoints = 25
pcargs.MaxStepSize = 0.1
pcargs.LocBifPoints = 'all'
pcargs.SaveEigen = True
pc.newCurve(pcargs)

# Plot bifurcation diagram
pc.display(('k', 'D'), stability=True)
plt.title('Bifurcation diagram of equilibria in (k,D) parameters')

# D value increases monotonically with k, so let's find the value at k=9.0
# by linear interpolation from the points found by PyCont
hopfs_unparam = pc['HO1'].sol[['D', 'v1', 'v2', 'p1', 'p2']]
hopfs = dst.Pointset(indepvararray=pc['HO1'].sol['k'], indepvarname='k', coorddict=hopfs_unparam.todict())
ix0, ix1 = hopfs.find(9)
print "k values found around 9.0 are %.3f, %.3f" % (hopfs['k'][ix0], hopfs['k'][ix1])

# Arbitrarily choose closest lower index to check stability
print "Hopf point found for k = %.3f is stable?" % hopfs['k'][ix0],
if pc['HO1'].sol[ix0].labels['H']['stab'] == 'N':
    print "No"
else:
    print "Yes"

# Create interpolatable curve from the pointset
hopfs_curve = dst.pointset_to_traj(hopfs)
# Interpolate unstable Hopf equilibrium values of parameters/variables at k=9
hpt = hopfs_curve(9.0)

# extract the variables defined for 'ode' Generator
hics = hpt[ode.funcspec.vars]
# perturb slightly off equilibrium
hics['v2'] += 0.01

# extract the parameter value for D and indicate on bifurcation diagram
plt.plot(9.0, hpt['D'], 'go')
```

# Noisy Hodgkin-Huxley model



```python
from PyDSTool.Toolbox import neuralcomp

# voltage is already defined as the string 'V'
v = dst.Var(neuralcomp.voltage)

NaF = dst.args()
NaF.g = dst.Par(100, 'g')
NaF.Erev = dst.Par(50, 'vrev')
NaF.m = dst.Var('m')
NaF.h = dst.Var('h')

# Here are the forward and backward rate equations for each gate
# These are QuantSpec types that store the expression as a string
NaF.malpha = 0.32 * (v + 54) / (1 - dst.Exp(-(v + 54) / 4))
NaF.mbeta = 0.28 * (v + 27) / (dst.Exp((v + 27) / 5) - 1)
NaF.halpha = .128 * dst.Exp(-(50 + v) / 18)
NaF.hbeta = 4 / (1 + dst.Exp(-(v + 27) / 5))

# defining RHS expressions with a name attribute that matches a Variable creates an ODE for it
NaF_m_RHS = dst.Var(NaF.malpha * (1 - NaF.m) - NaF.mbeta * NaF.m, name=NaF.m.name, domain=[0, 1], specType='RHSfuncSpec')
NaF_h_RHS = dst.Var(NaF.halpha * (1 - NaF.h) - NaF.hbeta * NaF.h, name=NaF.h.name, domain=[0, 1], specType='RHSfuncSpec')

# otherwise, you can write explicit function specifications
# the Var 'I' is reserved by the channel object for the channel current
NaF.I = dst.Var(dst.Pow(NaF.m, 3) * NaF.h * NaF.g * (v - NaF.Erev), name='I', specType='ExpFuncSpec')

K = dst.args()
K.g = dst.Par(80, 'g')
K.Erev = dst.Par(-100, 'vrev')
K.n = dst.Var('n')
K.nalpha = .032 * (v + 52) / (1 - dst.Exp(-(v + 52) / 5))
K.nbeta = .5 * dst.Exp(-(57 + v) / 40)
K_n_RHS = dst.Var(K.nalpha * (1 - K.n) - K.nbeta * K.n, name=K.n.name, domain=[0, 1], specType='RHSfuncSpec')
K.I = dst.Var(dst.Pow(K.n, 4) * K.g * (v - K.Erev), name='I', specType='ExpFuncSpec')

L = dst.args()
L.g = dst.Par(0.1, 'g')
L.Erev = dst.Par(-67, 'vrev')
L.I = dst.Var(L.g * (v - L.Erev), name='I', specType='ExpFuncSpec')

channel_NaF = neuralcomp.channel('NaF')  # at minimum, a new channel needs a name
channel_NaF.add([NaF.g,
                 NaF.Erev,
                 NaF_m_RHS,
                 NaF_h_RHS,
                 NaF.I
                 ])

channel_K = neuralcomp.channel('K')
channel_K.add([K.g,
               K.Erev,
               K_n_RHS,
               K.I
               ])

channel_L = neuralcomp.channel('L')
channel_L.add([L.g,
               L.Erev,
               L.I
               ])

channel_bias = neuralcomp.channel('Ibias')
Ibias = dst.Par(1.75, 'Ibias')
Ib = dst.Var(-Ibias, name='I', specType='ExpFuncSpec')
channel_bias.add([Ibias, Ib])

time = dst.Var('t')
channel_ext = neuralcomp.channel('Iext')
mag_ext = dst.Par(3, 'Iext_magfac')
Iext_func = dst.Fun(mag_ext * 0.5 * (dst.Sin(2.124 * time) + dst.Sin(5.1 * (time + 0.2)))
                    * dst.Cos(3 * dst.pi * (time - 0.1)) * dst.Min(dst.Max(dst.Tan(8.982 * time * dst.pi + 1.3005), 8), -8),
                    ('t',), 'Iext_func')
Iext = dst.Var(-Iext_func(time), name='I', specType='ExpFuncSpec')
channel_ext.add([mag_ext, Iext_func, Iext])

cell1 = neuralcomp.makeSoma('cell1',channelList=[channel_NaF, channel_K, channel_L, channel_bias, channel_ext], C=1) # capacitance

alg_args = {'init_step': 0.03}
#targetGen = 'Radau_ODEsystem'
#targetGen = 'Dopri_ODEsystem'
targetGen = 'Vode_ODEsystem'


ic_args_net = {'V': -70.0,
               'NaF.m': .2,
               'NaF.h': .8,
               'K.n': .2}

model_HH = neuralcomp.ModelConstructor(
    'HH_model', generatorspecs={cell1.name: {'modelspec': cell1,
                                             'target': targetGen,
                                             'algparams': alg_args}},
    indepvar=('t', [0,100]))

HH_model = model_HH.getModel()

verboselevel = 2
print "Computing trajectory using verbosity level %d..." % verboselevel

HH_model.compute(trajname='test',
                 tdata=[0, 100],
                 ics=ic_args_net,
                 verboselevel=verboselevel)

v_dat = HH_model.sample('test')

plt.plot(v_dat['t'], v_dat['V'], 'b', label="Ibias=1.75")

HH_model.setPars('Ibias.Ibias', 2.75)
HH_model.compute(trajname='test2',
                 tdata=[0, 10000],
                 ics=ic_args_net,
                 verboselevel=verboselevel)

v_dat2 = HH_model.sample('test2')
plt.plot(v_dat2['t'], v_dat2['V'], 'r', label="Ibias=2.75")
plt.xlabel("t")
plt.ylabel("V")
plt.legend()
```
